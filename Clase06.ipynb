```python
my_dict = {"Size":["L", "L", "M", "M", "M", "S", "S", "S", "S"],
           "Type":["A", "A", "A", "B", "B", "A", "A", "B", "B"],
           "Location":["L1", "L1", "L1", "L2", "L2", "L1", "L2", "L2", "L1"],
           "A":[1, 2, 2, 3, 3, 4, 5, 6, 7],
           "B":[2, 4, 5, 5, 6, 6, 8, 9, 9]}
df = pd.DataFrame(my_dict)
df
```

```python
dfr = pd.pivot_table(df, index=['Size','Type'], columns='Location', values='B')
```

```python
dfr
```

```python
dfr.sum(axis=0)
```

```python
dfr.sum(axis=1)
```

```python
dfr.mean(axis=0, skipna=False)
```

```python
dfr.mean(axis=0, skipna=True)
```
